```javascript
// Solution Q.4 How to print the first non-repeated character in a string?

str1="abcd";
str_char=[];
for(i=0;i<str1.length-1;i++){
    rep=0;
    for(j=i+1;j<str1.length;j++){
        
        if (str_char.length>0){
        if(str_char.includes(str1[i])){
    
            
            rep++;
            break;
        }
        }
        if(str1[i]!==str1[j]){
        
            continue;
        }
        else
        {
            
            rep+=1;
            break;
        }
        }
        
        
        if(rep==0)
        {
        
        console.log("The first non-repeated character of the string is: " + str1[i]);
        break;
        
        }
        
        str_char.push(str1[i]);
        
}

if(rep!==0){
    console.log("There is no character in the string: " + str1 + " which is not repeated.")
}
```

```javascript
// Solution Q.5 Implementation of algorithm of Tower of Hanoi.


function towerofHanoi(n,from,to,aux){
    if(n==1){
        console.log(`Move disk ${n} from ${from} rod to ${to} rod`);
        return;
    }
    
    towerofHanoi(n-1,from,aux,to);
    
//     function calling itself until the no. of disks reduce to 1 (recursion)
    
    console.log(`Move disk ${n} from ${from} rod to ${to} rod`);
    towerofHanoi(n-1,aux,to,from);
    
}

towerofHanoi(3,'A','C','B') 

// no. of disks : 3, source pole : A, destination pole: C , auxillary pole: B
```

```javascript
// Solution: Q.6 How to convert postfix expression to prefix expression?

// function to check if scanned val is operator 
function isOperator(x){
    
    if(x=='+' || x=='-' || x=='*' || x=='/' || x=='%' || x=='^')
    
    return true;
    
    else
    
    return false;
}

//function to convert postfix expression into prefix expression
function post_to_prefix(expr){
    
    let st=[];
    
    //loop for reading values from left to right
    
    for(i=0;i<expr.length;i++){
        
        let scan_val=expr[i];
        
        if (isOperator(scan_val)){
            
            let operand2=st.pop();
            
            let operand1=st.pop();
            
            let temp= scan_val + operand1 + operand2;
            
            st.push(temp);
            
        }
        
        else{
            
            st.push(scan_val);
        }
    }
    

    
    return st[st.length-1];
    
}

let ex="ABC/-AK/L-*";
console.log(post_to_prefix(ex));

```

```javascript
// Solution: Q.6 How to convert prefix expression to infix expression?

// function to check if scanned val is operator 
function isOperator(x){
    
    if(x=='+' || x=='-' || x=='*' || x=='/' || x=='%' || x=='^')
    
    return true;
    
    else
    
    return false;
}


// function to convert prefix expression into infix expression

function pre_to_infix(expr){
    
    let st=[];
    
    //loop for reading values from right to left
    
    for(i=expr.length-1;i>=0;i--){
        
        let scan_val=expr[i];
        
        if (isOperator(scan_val)){
            
            let operand1=st.pop();
            
            let operand2=st.pop();
            
            let temp= "(" + operand1 + scan_val + operand2 + ")";
            
            st.push(temp);
            
        }
        
        else{
            
            st.push(scan_val);
        }
    }
    
    return st[st.length-1];
    
}

let e = "*+AB-CD";
console.log(pre_to_infix(e));
```

```javascript
// Soution: Q.7 Program to check if all the brackets in the given snippet are closed.

function br_bal(expr){
    let br_st=[];
    
    
    let count=0;
    for(i=0;i<expr.length;i++){
        
        let x=expr[i];
        
        // if any opening bracket is found, push it in the brackets stack
        
        if(x=='(' || x=='{' || x=='['){
            
            // count var to denote if either(opening or closing) of any kind of brackets exists in the snippet
            
            count++;
            br_st.push(x);
            
            
        }
        
        let brkts_in_st;
        
        // in case where any closing bracket is scanned, 
        // the bracket present at the top of the stack should match 
        // with the type of scanned bracket, else return false
        
        switch(x){
            case ')':
                
                count++;
                
                if (br_st.length!==0){
                brkts_in_st=br_st.pop();
                
                if(brkts_in_st=='('){
                    
                    continue;
                }
                else{
                    return false;
                }
                }
                else
                
                return false;
                
                break;
                
           case '}':
               
               count++;
               
               if(br_st!==0){
               brkts_in_st=br_st.pop();
               
               if(brkts_in_st=='{'){
                   continue;
               }
               else{
                   return false;
               }
               }
               else
               return false;
               
               break;
               
           case ']':
               
               count++;
             if(br_st.length!==0){
                
               brkts_in_st=br_st.pop();
               
               if(brkts_in_st=='[')
                   
                   continue;
                   
              else
                
                return false;
                
                 }
        
            else 
        
                return false;
            
            break;
        }
        
    }
    
    if (count==0){
        
        str='No Bracket Found'
        return str;
    }
    
    return (br_st.length==0);
}

str="(Hello";
console.log("Are the brackets in the expression balanced: " + br_bal(str));

```

```javascript
// Solution: Q.9 how to reverse a stack?

class Stack{
    
    constructor(){
        this.st=[]; // stack for storing values 
        
    }
    
    // method to insert values in stack
    
    push(val){
        
        this.st.push(val);
        
    }
    
    pop(){
        
        
        // if the stack is empty, return 'underflow'
        if (this.st.length==0)
        
        return 'underflow';
        
        else
            
            // returning the value popped from the stack
            return this.st.pop();
        
        
    }
    
    peek(){
        
        // returning the value present at the last index position
        
        return this.st[this.st.length-1];
    }
    
    isEmpty(){
        
        // returning true, if the length of stack is 0, else false
        
        return this.st.length==0;
    }
    
    
    size(){
        
        // returning the length of the stack
        
        return this.st.length;
    }
    
    // stack method to reverse stack
    
    reverse(){
        
        if (this.st.length>0){
            
            //holding all values in the callback function until the stack is empty
            
            let x=this.st.pop();
            
            //the function is calling itself until there is no value left in the stack
            
            this.reverse();
            
            // function call for inserting values held in the callback function at the bottom of the stack
            
            this.insert_at_bottom(x);
            
        }
        
    }
        
        
        // stack method to push value at the bottom of the stack
        
        insert_at_bottom(x){
            
            if (this.st.length==0)
                
                this.push(x);
                
            else{
                
                // all the values of the stack are held in callback 
                //function until the stack is empty, 
                //and then value in the function call is inserted at the bottom of the stack
                let a=this.st.pop();
                
                this.insert_at_bottom(x);
                
                // push items popped from the stack ,after the value in function call is pushed at the bottom of the stack
                
                this.push(a);
            }
                
    
        }
        
        printStack(){
            
            var str="";
            for(let i=0;i<this.st.length;i++){
                
                str+=this.st[i] + " ";
            }
            
            return str;
        }
    
}

var stk=new Stack();

stk.push(100);
stk.push(50);
stk.push(30);

stk.push(95);

stk.push(65);


console.log(stk.printStack());

stk.reverse();

console.log(stk.printStack());
```

```javascript
// Solution: Q.10 How to get minimum value in the stack?

// stack class which returns min value of stack in O(1) time 

class Stack_{
    
    constructor(){
        this.st=[]; // stack for storing values 
        
        this.min_st=[]; //stack for storing min value in the existing stack
        
    }
    
    // method to insert values in stack
    
    push(val){
        
        // if stack is empty, simply insert value in both the stack  & min_stack
        
        if(this.st.length==0){
            
            this.st.push(val);
            this.min_st.push(val);
        }
        
        else{
            
            // if stack is not empty,insert value in the stack
            this.st.push(val);
            
            // if stack is not empty, 
            //compare the existing value in the min_stack 
            //with the value to be inserted in the stack & insert the minimum of both the values
            
            this.min_st.push(Math.min(this.min_st[this.min_st.length-1],val));
        }
        
    }
    
    remove(){
        
        
        // return 'underflow', if the stack is empty
        if (this.st.length==0)
        
        return 'underflow';
        
        else{
            
            // pop the value from the min_stack
            this.min_st.pop();
            
            // returning the value popped from the stack
            return this.st.pop();
        }
        
    }
    
    peek(){
        
        // returning the value present at the last index position
        
        return this.st[this.st.length-1];
    }
    
    isEmpty(){
        
        // returning true, if the length of stack is 0, else false
        
        return this.st.length==0;
    }
    
    
    size(){
        
        // returning the length of the stack
        
        return this.st.length;
    }
    
    min(){
        
        // returning the minimum value of the stack, which is maintained at the last index position of the min_stack
        
        return this.min_st[this.min_st.length-1];
    }
    
}

var stk=new Stack_();

stk.push(100);
stk.push(50);
stk.push(30);

console.log(stk.min());

```
